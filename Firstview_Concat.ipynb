```python
import pandas as pd
```

```python
red = pd.read_csv('data/winequality-red.csv', sep = ';')
white = pd.read_csv('data/winequality-white.csv', sep = ';')
red['type'] = ['red' for _ in range(len(red))]
white['type'] = ['white' for _ in range(len(white))]
```

```python
red.columns == white.columns
```

```python
all =  pd.concat([red, white])
all = all.sample(frac=1, random_state=42)
```

```python
all.to_csv('data/whinequality.csv', index=False)
```
